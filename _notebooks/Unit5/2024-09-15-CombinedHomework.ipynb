```java
import java.util.ArrayList;
import java.util.List;

public class Villain {
    // Instance variables
    public String name;
    public String evilPlan;
    public List<String> minions;
    public static int villainCount = 0;

    // Static list to keep track of all minions across all villains
    private static List<String> globalMinionList = new ArrayList<>();

    // Constructor for name, plan, and minions
    public Villain(String name, String evilPlan) {
        this.name = name;
        this.evilPlan = evilPlan;
        this.minions = new ArrayList<>();
        villainCount++;
    }

    // Instance method to add a minion
    public void addMinion(String minion) {
        minions.add(minion);
        globalMinionList.add(minion);  // Add to global list of minions
        System.out.println(minion + " has been added to " + name + "'s army.");
    }

    // Instance method to describe the villain
    public void describeVillain() {
        System.out.println(name + " is planning to: " + evilPlan);
        System.out.println("They have " + minions.size() + " minions.");
    }

    // Static method to get the total count of villains
    public static int getVillainCount() {
        return villainCount;
    }

    // New static method to print all minions across all villains
    public static void printAllMinions() {
        System.out.println("=== Global Minion List ===");
        if (globalMinionList.isEmpty()) {
            System.out.println("No minions found.");
        } else {
            for (String minion : globalMinionList) {
                System.out.println("Minion: " + minion);
            }
        }
        System.out.println("==========================");
    }
}

public class Main {
    public static void main(String[] args) {
        Villain.villainCount = 0;

        // Create new villains
        Villain gru = new Villain("Gru", "steal the moon!");
        Villain vector = new Villain("Vector", "take over the world with magnitude and direction!");

        System.out.println("=== Adding Minions ===");
        // Create some minions
        gru.addMinion("Kevin");
        gru.addMinion("Stuart");
        gru.addMinion("Bob");

        // Create some minions for Vector
        vector.addMinion("Henchman 1");
        vector.addMinion("Henchman 2");

        System.out.println();

        // Describe the villains and their plans
        System.out.println("=== Villain Descriptions ===");
        gru.describeVillain();
        System.out.println();
        vector.describeVillain();
        System.out.println();

        // Get the total count of villains
        System.out.println("=== Total Villain Count ===");
        System.out.println("There are " + Villain.getVillainCount() + " villains in the world.");

        // Call the new static method to print all minions
        Villain.printAllMinions();
    }
}
Main.main(null);
```

```java
public class Minion {
    private String color;
    private String name;
    private int energyLevels;
    private String gadgets;
    private String hair;
    private int height;
    public static int totalEnergyLevels = 0;
    public static int minionCount = 0;

    public Minion(String color, String name, int energyLevels, String gadgets, String hair, int height) {
        this.color = color;
        this.name = name;
        this.energyLevels = energyLevels;
        this.gadgets = gadgets;
        this.hair = hair;
        this.height = height;
        totalEnergyLevels += energyLevels;
        minionCount++;
    }

    public static double getAverageEnergyLevels() {
        return (double) totalEnergyLevels / minionCount;
    }
}

public class Main {
    public static void main(String[] args) {
        Minion minion1 = new Minion("Pink", "Kevin", 80, "Banana Blaster", "Tall", 100);
        Minion minion2 = new Minion("Purple", "Stuart", 10, "Big Car", "Bald", 145);
        Minion minion3 = new Minion("Red", "Bob", 1000, "Bubble Gum Blaster", "Curly", 160);
        
        System.out.println("Average Energy Levels: " + Minion.getAverageEnergyLevels());
    }
}
Main.main(null);
```

```java
import java.util.ArrayList;
import java.util.List;
import java.util.Scanner;

// Base Minion class
class Minion {
    private String name;
    private String color;
    private int energyLevel;
    private String task;
    private String gadget;

    // Static variable to keep track of all minions created
    public static List<Minion> allMinions = new ArrayList<>();

    // Constructor to initialize Minion attributes
    public Minion(String name, String color, int energyLevel, String task, String gadget) {
        this.name = name;
        this.color = color;
        this.energyLevel = energyLevel;
        this.task = task;
        this.gadget = gadget;
        allMinions.add(this);  // Add this minion to the global list
    }

    // Method to calculate recovery time based on color and energy level
    public int calculateRecoveryTime() {
        int baseRecoveryTime;
        if (color.equalsIgnoreCase("Purple")) {
            baseRecoveryTime = 100 - energyLevel;
        } else {
            baseRecoveryTime = 50 - energyLevel;
        }
        if (energyLevel < 30) {
            baseRecoveryTime += 20;
        }
        return baseRecoveryTime;
    }

    // Getter methods for Minion attributes
    public String getName() { return name; }
    public String getColor() { return color; }
    public int getEnergyLevel() { return energyLevel; }
    public String getTask() { return task; }
    public String getGadget() { return gadget; }

    // Display minion details
    public void displayDetails() {
        System.out.println("Minion Name: " + name);
        System.out.println("Color: " + color);
        System.out.println("Energy Level: " + energyLevel);
        System.out.println("Task: " + task);
        System.out.println("Gadget: " + gadget);
        System.out.println("Recovery Time: " + calculateRecoveryTime() + " hours\n");
    }
}

// Specialized Minions
class Kevin extends Minion {
    public Kevin() {
        super("Kevin", "Yellow", 80, "Leader", "Banana Gun");
    }
}

class Stuart extends Minion {
    public Stuart() {
        super("Stuart", "Yellow", 60, "Lookout", "Guitar");
    }
}

class Bob extends Minion {
    public Bob() {
        super("Bob", "Purple", 40, "Distraction", "Plush Teddy");
    }
}

// Gadget class to handle gadget costs
class Gadget {
    private String name;
    private double cost;

    // Static variable to track the total cost of all gadgets
    private static double totalGadgetCost = 0;

    // Constructor to initialize gadget name and cost
    public Gadget(String name, double cost) {
        this.name = name;
        this.cost = cost;
        totalGadgetCost += cost;  // Add to total cost
    }

    // Static method to get the total gadget cost
    public static double getTotalGadgetCost() {
        return totalGadgetCost;
    }
}

// Game class to run the minion game
public class MinionGame {
    public static void main(String[] args) {
        // Create Minion objects
        Kevin kevin = new Kevin();
        Stuart stuart = new Stuart();
        Bob bob = new Bob();

        // Create Gadget objects
        Gadget bananaGun = new Gadget("Banana Gun", 500);
        Gadget freezeRay = new Gadget("Freeze Ray", 1000);
        Gadget rocketLauncher = new Gadget("Rocket Launcher", 1500);

        // Scanner for user input
        Scanner scanner = new Scanner(System.in);
        boolean exit = false;

        // Game Menu Loop
        while (!exit) {
            System.out.println("=== Minion Management Game ===");
            System.out.println("1. Display All Minions");
            System.out.println("2. Display Total Gadget Cost");
            System.out.println("3. Add a New Minion");
            System.out.println("4. Exit");
            System.out.print("Enter your choice: ");
            int choice = scanner.nextInt();

            switch (choice) {
                case 1:
                    System.out.println("\n=== All Minion Details ===");
                    for (Minion minion : Minion.allMinions) {
                        minion.displayDetails();
                    }
                    break;

                case 2:
                    System.out.printf("Total Cost of All Gadgets: $%.2f\n\n", Gadget.getTotalGadgetCost());
                    break;

                case 3:
                    // Add a new minion
                    scanner.nextLine();  // Consume newline
                    System.out.print("Enter Minion Name: ");
                    String name = scanner.nextLine();
                    System.out.print("Enter Minion Color (Yellow/Purple): ");
                    String color = scanner.nextLine();
                    System.out.print("Enter Energy Level (1-100): ");
                    int energyLevel = scanner.nextInt();
                    scanner.nextLine();  // Consume newline
                    System.out.print("Enter Minion Task: ");
                    String task = scanner.nextLine();
                    System.out.print("Enter Minion Gadget: ");
                    String gadget = scanner.nextLine();

                    Minion newMinion = new Minion(name, color, energyLevel, task, gadget);
                    System.out.println("New Minion " + name + " has been added!\n");
                    break;

                case 4:
                    exit = true;
                    System.out.println("Exiting the game. Bye!");
                    break;

                default:
                    System.out.println("Invalid choice! Please try again.\n");
            }
        }

        scanner.close();
    }
}
MinionGame.main(null);
```
