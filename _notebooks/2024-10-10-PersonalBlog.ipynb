Polymorphism & Superclasses
Polymorphism = Using the same method name but having different behavior based on the object type.

Example in Code:
Parent Class: Car has a displayInfo() method.
Subclasses: SportsCar, SUV, Truck override displayInfo().
Each subclass adds its own details when displayInfo() is called.
Why Polymorphism?
Lets subclasses act differently while sharing the same method name.
You can treat a SportsCar, SUV, or Truck as just a Car but still get the specific behavior of each type.
super Keyword
Used in constructors to call the parent class and set shared values (make, model, etc.).

```java
import java.util.Scanner;
class Car {
    String make;
    String model;
    int year;
    public Car(String make, String model, int year) {
        this.make = make;
        this.model = model;
        this.year = year;
    }
    public void displayInfo() {
        System.out.println("Car Details:");
        System.out.println("Make: " + make);
        System.out.println("Model: " + model);
        System.out.println("Year: " + year);
    }
}
class SportsCar extends Car {
    String topSpeed;
    public SportsCar(String make, String model, int year, String topSpeed) {
        super(make, model, year); 
        this.topSpeed = topSpeed;
    }

    @Override
    public void displayInfo() {
        super.displayInfo();
        System.out.println("Type: Sports Car");
        System.out.println("Top Speed: " + topSpeed);
    }
}

// Subclass representing an SUV
class SUV extends Car {
    boolean isFourWheelDrive;

    // Constructor using the super keyword to call the parent constructor
    public SUV(String make, String model, int year, boolean isFourWheelDrive) {
        super(make, model, year);  // Call the base Car constructor
        this.isFourWheelDrive = isFourWheelDrive;
    }

    @Override
    public void displayInfo() {
        super.displayInfo();
        System.out.println("Type: SUV");
        System.out.println("Four-Wheel Drive: " + (isFourWheelDrive ? "Yes" : "No"));
    }
}

// Subclass representing a Truck
class Truck extends Car {
    double towingCapacity;

    // Constructor using the super keyword to call the parent constructor
    public Truck(String make, String model, int year, double towingCapacity) {
        super(make, model, year);  // Call the base Car constructor
        this.towingCapacity = towingCapacity;
    }

    @Override
    public void displayInfo() {
        super.displayInfo();
        System.out.println("Type: Truck");
        System.out.println("Towing Capacity: " + towingCapacity + " tons");
    }
}

public class CarGame {
    public static void main(String[] args) {
        Scanner input = new Scanner(System.in);
        boolean exit = false;

        while (!exit) {
            System.out.println("\n--- Welcome to the Car Game! ---");
            System.out.println("Choose the type of vehicle you want to create:");
            System.out.println("1. Sports Car");
            System.out.println("2. SUV");
            System.out.println("3. Truck");
            System.out.println("4. Exit");
            System.out.print("Enter your choice: ");
            
            int choice = input.nextInt();
            input.nextLine();  // Clear the buffer
            
            switch (choice) {
                case 1:  // Create a Sports Car
                    System.out.print("Enter the make of the car: ");
                    String sportsMake = input.nextLine();
                    System.out.print("Enter the model of the car: ");
                    String sportsModel = input.nextLine();
                    System.out.print("Enter the year of the car: ");
                    int sportsYear = input.nextInt();
                    input.nextLine();  // Clear the buffer
                    System.out.print("Enter the top speed of the car (e.g., 200 mph): ");
                    String topSpeed = input.nextLine();
                    
                    SportsCar sportsCar = new SportsCar(sportsMake, sportsModel, sportsYear, topSpeed);
                    System.out.println("\n--- Your Sports Car ---");
                    sportsCar.displayInfo();
                    break;
                    
                case 2:  // Create an SUV
                    System.out.print("Enter the make of the SUV: ");
                    String suvMake = input.nextLine();
                    System.out.print("Enter the model of the SUV: ");
                    String suvModel = input.nextLine();
                    System.out.print("Enter the year of the SUV: ");
                    int suvYear = input.nextInt();
                    System.out.print("Does the SUV have four-wheel drive? (true/false): ");
                    boolean isFourWheelDrive = input.nextBoolean();
                    
                    SUV suv = new SUV(suvMake, suvModel, suvYear, isFourWheelDrive);
                    System.out.println("\n--- Your SUV ---");
                    suv.displayInfo();
                    break;
                    
                case 3:  // Create a Truck
                    System.out.print("Enter the make of the truck: ");
                    String truckMake = input.nextLine();
                    System.out.print("Enter the model of the truck: ");
                    String truckModel = input.nextLine();
                    System.out.print("Enter the year of the truck: ");
                    int truckYear = input.nextInt();
                    System.out.print("Enter the towing capacity in tons (e.g., 5.5): ");
                    double towingCapacity = input.nextDouble();
                    
                    Truck truck = new Truck(truckMake, truckModel, truckYear, towingCapacity);
                    System.out.println("\n--- Your Truck ---");
                    truck.displayInfo();
                    break;
                    
                case 4:  // Exit the game
                    exit = true;
                    System.out.println("Exiting the Car Game. Goodbye!");
                    break;
                    
                default:
                    System.out.println("Invalid choice. Please try again.");
            }
        }
        input.close();
    }
}
CarGame.main(null);
```
